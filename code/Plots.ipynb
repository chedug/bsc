```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate some sample data
np.random.seed(8)
X = 2 * np.random.rand(50, 1)
y = 4 + 3 * X + np.random.randn(50, 1)

# Fit a linear regression model
model = LinearRegression()
model.fit(X, y)

# Predicted values
y_pred = model.predict(X)

# Select two random points
index1, index2 = np.random.choice(len(X), 2, replace=False)

# Create a zoomed-in version around the selected point
zoom_factor = 1.2
xmin, xmax = X[index1] - zoom_factor, X[index1] + zoom_factor
ymin, ymax = min(y[index1], y_pred[index1]) - zoom_factor, max(y[index1], y_pred[index1]) + zoom_factor

# Plot the scatter plot
plt.scatter(X, y, label='Observed Data')

# Plot the linear regression line
plt.plot(X, y_pred, color='black', linewidth=1, label='Linear Regression Line')

# Plot the observed point (y_i)
plt.scatter(X[index1], y[index1], color='green', label='$y_i$')

# Plot the fitted point (\hat{y_i})
plt.scatter(X[index1], y_pred[index1], color='black', label='$\hat{y_i}$')

# Connect the points with a vertical line representing the error term (\epsilon_i)

plt.plot([X[index1], X[index1]], [y[index1], y_pred[index1]], color='black', linestyle='dashed', linewidth=1)

# Set axis limits for the zoomed-in version


# Add labels and legend
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (Y)')
plt.legend()

# Show the plot
plt.show()
```
